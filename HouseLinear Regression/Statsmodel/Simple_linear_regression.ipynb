```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_boston = pd.read_csv("Boston House Prices.csv")
```

```python
df_boston
```

```python
# define the dependent and independent values
x = df_boston["Rooms"] # independet
y = df_boston["Value"] # dependent
```

```python
# explore the dataframe on scatterplot
df_boston.plot(kind="scatter", x="Rooms", y="Value", 
               figsize=(8,4))
plt.show()
```

```python
# adding a constant and fit the model
x = sm.add_constant(x) # constant
lm = sm.OLS(y,x).fit() # fitting the model
```

```python
lm.predict(x)
```

```python
# regression table
lm.summary()
```

```python
# linear equation: y = ax + b
y_pred = 9.1021*x["Rooms"] - 34.6706	
```

```python
# plotting the line regression line
plt.figure(figsize=(10,5), tight_layout=True)
sns.scatterplot(x=x["Rooms"], y=y)
sns.lineplot(x=x["Rooms"], y=y_pred, color="red")
plt.xlim(0)
plt.ylim(0)
plt.show()
```
