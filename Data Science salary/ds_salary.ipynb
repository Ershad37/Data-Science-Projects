```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_salary_raw = pd.read_csv("ds_salaries.csv")
```

```python
df_salary = df_salary_raw.copy(deep=True)
```

```python
# check if the dataframe has null values
df_salary.isnull().sum()
```

```python
# drop some columns
df_salary.drop(columns=["salary", "salary_currency"], axis=1, inplace=True)
```

```python
# check for data type
df_salary.info()
```

```python
# checks for value counts of each column
for col in df_salary.columns:
    print(f"The value counts for {col}")
    print(df_salary[col].value_counts())
    print()
```

```python
# rename the values in columns
df_salary["experience_level"] = df_salary["experience_level"].replace("SE", "Expert")
df_salary["experience_level"] = df_salary["experience_level"].replace("EN", "Junior")
df_salary["experience_level"] = df_salary["experience_level"].replace("MI", "Intermediate")
df_salary["experience_level"] = df_salary["experience_level"].replace("EX", "Director")

df_salary["employment_type"] = df_salary["employment_type"].replace("FT", "Full_time")
df_salary["employment_type"] = df_salary["employment_type"].replace("PT", "Part_time")
df_salary["employment_type"] = df_salary["employment_type"].replace("CT", "Contract")
df_salary["employment_type"] = df_salary["employment_type"].replace("FL", "Freelance")

df_salary["company_size"] = df_salary["company_size"].replace("S", "Small")
df_salary["company_size"] = df_salary["company_size"].replace("M", "Medium")
df_salary["company_size"] = df_salary["company_size"].replace("L", "Large")
```

```python
df_salary.value_counts("experience_level")
```

```python
counts = df_salary["experience_level"].value_counts()
counts.head().index
sns.countplot(data=df_salary, x="experience_level", legend=False, palette="rocket")
plt.show()
```
